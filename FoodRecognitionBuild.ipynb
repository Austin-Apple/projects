```python
import os
import cv2
import numpy as np
from numpy import load
from numpy.core.defchararray import index
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2" #reduces extra info messages
```

```python
class_names = ["banana", "apple", "pear", "grapes", "orange", "kiwi", "watermelon", "pomegranate", "pineapple", "mango", "cucumber", "carrot", "capsicum", "onion", "potato", "lemon", "tomato", "raddish", "beetroot", "cabbage", "lettuce", "spinach", "soy beans", "cauliflower", "bell pepper", "chilli pepper", "turnip", "corn", "sweetcorn", "sweetpotato", "paprika", "jalepeno", "ginger", "garlic", "peas", "eggplant"]
```

```python
#load in all the datasets we collected
train_data = load('/Users/apple/jupyter/train_data.npy')
train_data_labels = load('/Users/apple/jupyter/train_data_labels.npy')

test_data = load('/Users/apple/jupyter/test_data.npy')
test_data_big = load('/Users/apple/jupyter/test_data_big.npy')
test_data_labels = load('/Users/apple/jupyter/test_data_labels.npy')
```

```python
# 3144 images 
print("train shape:", train_data.shape)
print("train labels shape:", train_data_labels.shape)
print("test data shape:", test_data.shape)
print("test data labels shape:", test_data_labels.shape)
```

```python
train_data = train_data / 255.0
test_data = test_data / 255.0
```

```python
model = keras.Sequential([
    #flattens image for input into the model
    keras.layers.Flatten(input_shape=(28,28,3)), 

    #hidden layer and we will activate 512 neurons
    keras.layers.Dense(512, activation='relu'),

    #last layer for the classification of classes
    keras.layers.Dense(36, activation='softmax')
])
```

```python
model.compile(
    #optimizer calculates the gradient descent of the netword
    optimizer = 'adam',
    #loss function
    loss = 'sparse_categorical_crossentropy',
    #metrics measurment
    metrics = ['accuracy']
)
```

```python
model.fit(train_data,train_data_labels,epochs=120)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_data_labels,verbose=1)
print("******************* Test accuracy:", test_acc)
```

```python
predictions = model.predict(test_data)
```

```python
class_index = np.argmax(predictions[200])
print(class_index)
```

```python
for predict, test_label in zip(predictions,test_data_labels):
    class_index = np.argmax(predict)
    class_name_predict = class_names[class_index]

    class_name_original = class_names[test_label]
    print("Predicted class: ",class_name_predict, "OG: ",  class_name_original)
```
