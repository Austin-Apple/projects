```python
import os
import cv2
import numpy as np
from numpy import save
from numpy.core.defchararray import index
```

```python
class_names = ["banana", "apple", "pear", "grapes", "orange", "kiwi", "watermelon", "pomegranate", "pineapple", "mango", "cucumber", "carrot", "capsicum", "onion", "potato", "lemon", "tomato", "raddish", "beetroot", "cabbage", "lettuce", "spinach", "soy beans", "cauliflower", "bell pepper", "chilli pepper", "turnip", "corn", "sweetcorn", "sweetpotato", "paprika", "jalepeno", "ginger", "garlic", "peas", "eggplant"]
```

```python
train_data_array = []
train_data_labels_array = []

test_data_array = []
test_data_labels_array = []
test_data_big_array = []
```

```python
rootdir = "/Users/apple/Downloads/fruitsveg/train" #grabs data set
```

```python
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        frame = cv2.imread(os.path.join(subdir,file)) #opens each image in the folders

        #if image is valid
        if frame is None:
            print("Not an image")
        else:
            #print(subdir,file)
            resized = cv2.resize(frame,(28,28), interpolation=cv2.INTER_AREA) #makes all the images the same size for consistency
            checkSize = resized.shape[0] #checks if resizing was successful
            if checkSize == 28:
                train_data_array.append(resized)
                i = class_names.index(os.path.basename(subdir))
                train_data_labels_array.append(i)

#convers the lists to numpy arrays
train_data = np.array(train_data_array)
train_data_labels = np.array(train_data_labels_array)

print("Finished loading the training data")
print("Num of training records:", train_data.shape[0])

print(train_data.shape)
print(train_data_labels.shape)
```

```python
#testing random image
demoImage = train_data[4] 
cv2.imshow("Demo image", demoImage)
index = train_data_labels[4]
print(class_names[index])
```

```python
save('/Users/apple/jupyter/train_data.npy', train_data)
save('/Users/apple/jupyter/train_data_labels.npy', train_data_labels)
```
